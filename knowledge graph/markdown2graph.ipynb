```python
def parse_markdown_to_json(md_text):
    """
    将形如:
        - Feature(心理特征)
            - Question(量表题目)
                - Scale(量表)
    的 Markdown 数据解析为:
        {
          Feature1: [
            {"question": "Question1", "scale": "Scale1"},
            {"question": "Question2", "scale": "Scale2"}
          ],
          Feature2: [
            ...
          ]
        }
    """
    # 按行拆分
    lines = md_text.splitlines()

    data = {}
    current_feature = None
    current_question = None

    for line in lines:
        # 将\t替换成4个空格（如果可能存在Tab）
        line = line.replace('\t', '    ')

        # 去掉行首行末多余空格，但保留缩进级别判断需要的信息
        # 这里先保留行首空格用于计算 indent
        raw_line = line

        # 如果整行只有空白，跳过
        if not raw_line.strip():
            continue

        # 仅处理以'-'开头的行
        if not raw_line.strip().startswith('-'):
            continue

        # 计算当前行前导空格数，以区分层级
        indent = len(raw_line) - len(raw_line.lstrip(' '))

        # 去掉行首的'- '部分，只保留后面的文本
        # 例如 raw_line = "    - 我觉得闷闷不乐" => line_stripped = "我觉得闷闷不乐"
        line_stripped = raw_line.lstrip(' ').lstrip('-').strip()

        if indent == 0:
            # 顶层 => 心理特征（Feature）
            current_feature = line_stripped
            data[current_feature] = []
            current_question = None

        elif indent == 4:
            # 第二层 => 量表题目（Question）
            current_question = line_stripped

        elif indent == 8:
            # 第三层 => 量表（Scale）
            scale = line_stripped
            # 将 {question, scale} 加入到当前 feature 对应的列表中
            if current_feature is not None and current_question is not None:
                data[current_feature].append({
                    "question": current_question,
                    "scale": scale
                })
            else:
                # 如果出现了缩进不匹配等异常情况，可在此处做相应错误处理
                pass

    return data

if __name__ == "__main__":
    # 将 Markdown 转换为 JSON
    parsed_result = parse_markdown_to_json(markdown_text)

    # 演示打印结果
    print(json.dumps(parsed_result, ensure_ascii=False, indent=2))
```

```python
from py2neo import Graph, Node, Relationship

# 根据自己的数据库配置进行修改
graph = Graph(
    "bolt://localhost:7687",  # 或者 "neo4j://localhost:7687"
    auth=("neo4j", "lc65120101")
)
```

```python
LABEL_FEATURE = "Feature"       # 心理特征
LABEL_QUESTION = "Question"     # 问题
LABEL_SCALE = "Scale"           # 量表
LABEL_OPTION = "Option"         # 选项

REL_HAS_QUESTION = "HAS_QUESTION"   # (Feature)-[:HAS_QUESTION]->(Question)
REL_BELONGS_TO = "BELONGS_TO"       # (Question)-[:BELONGS_TO]->(Scale)
REL_HAS_OPTION = "HAS_OPTION"       # (Question)-[:HAS_OPTION]->(Option)

# ================
# 4) 构建知识图谱
# ================
def build_knowledge_graph_from_parsed_result(data_dict):
    """
    data_dict 结构:
      {
        "心理特征A": [
          {
            "question": "题目1",
            "scale": "量表X",
            "options": [
              {"text": "没有"}, 
              {"text": "小部分时间"}, 
              ...
            ]
          },
          ...
        ],
        "心理特征B": [
          ...
        ]
      }
    """

    for feature_name, q_list in data_dict.items():
        # (1) 合并心理特征节点
        feature_node = Node(LABEL_FEATURE, name=feature_name)
        graph.merge(feature_node, LABEL_FEATURE, "name")

        # (2) 遍历该特征下的所有 question
        for q_item in q_list:
            question_text = q_item["question"]
            scale_name = q_item["scale"]
            options = q_item.get("options", [])  # 取出options列表

            # 合并 Question 节点
            question_node = Node(LABEL_QUESTION, text=question_text)
            graph.merge(question_node, LABEL_QUESTION, "text")

            # 合并 Scale 节点
            scale_node = Node(LABEL_SCALE, name=scale_name)
            graph.merge(scale_node, LABEL_SCALE, "name")

            # 创建 (Feature)-[:HAS_QUESTION]->(Question) 关系
            feature_question_rel = Relationship(feature_node, REL_HAS_QUESTION, question_node)
            graph.merge(feature_question_rel)

            # 创建 (Question)-[:BELONGS_TO]->(Scale) 关系
            question_scale_rel = Relationship(question_node, REL_BELONGS_TO, scale_node)
            graph.merge(question_scale_rel)

            # (3) 为该 Question 创建4个独立的选项节点
            for opt in options:
                opt_text = opt["text"]

                # 每个 Question 的选项都用 create 或 merge 的方式生成
                # 如果你想“不可共享”，就不要单纯用 text 做唯一属性，
                # 因为同样是 "没有" 也会 merge 到同一个节点中。
                # 可以使用 create 或者给它生成一个独特属性（如 question_id+option_text）。
                # 这里简单用 create，保证不会跟别的题目合并:
                option_node = Node(LABEL_OPTION, text=opt_text)

                # 创建节点
                graph.create(option_node)

                # 关系 (Question)-[:HAS_OPTION]->(Option)
                question_option_rel = Relationship(question_node, REL_HAS_OPTION, option_node)
                graph.create(question_option_rel)

                # 这样就确保每个 Question 都拥有自己的独立选项节点

if __name__ == "__main__":
    # 调用构建函数
    build_knowledge_graph_from_parsed_result(parsed_result)
    print("知识图谱构建完成！")
```

```python
LABEL_FEATURE = "Feature"     # 心理特征
LABEL_QUESTION = "Question"   # 量表题目
LABEL_SCALE = "Scale"         # 量表

REL_HAS_QUESTION = "HAS_QUESTION"  # (Feature)-[:HAS_QUESTION]->(Question)
REL_BELONGS_TO = "BELONGS_TO"      # (Question)-[:BELONGS_TO]->(Scale)

# =========================
# 4) 构建知识图谱函数
# =========================
def build_knowledge_graph_from_parsed_result(data_dict):
    """
    将形如:
    {
        "心理特征A": [
            {"question": "题目1", "scale": "量表X"},
            {"question": "题目2", "scale": "量表Y"}
        ],
        "心理特征B": [
            ...
        ]
    }
    的字典数据写入 Neo4j 知识图谱。
    """

    for feature_name, q_list in data_dict.items():
        # 1) 合并心理特征节点
        feature_node = Node(LABEL_FEATURE, name=feature_name)
        graph.merge(feature_node, LABEL_FEATURE, "name")

        # 2) 遍历该特征下的所有 {question, scale}
        for item in q_list:
            question_text = item["question"]
            scale_name = item["scale"]

            # 合并量表题目节点
            question_node = Node(LABEL_QUESTION, text=question_text)
            graph.merge(question_node, LABEL_QUESTION, "text")

            # 合并量表节点
            scale_node = Node(LABEL_SCALE, name=scale_name)
            graph.merge(scale_node, LABEL_SCALE, "name")

            # 3) 合并(Feature)-[:HAS_QUESTION]->(Question) 关系
            #   先节点后关系，避免重复创建
            feature_question_rel = Relationship(feature_node, REL_HAS_QUESTION, question_node)
            graph.merge(feature_question_rel)

            # 4) 合并(Question)-[:BELONGS_TO]->(Scale) 关系
            question_scale_rel = Relationship(question_node, REL_BELONGS_TO, scale_node)
            graph.merge(question_scale_rel)

if __name__ == "__main__":
    # =========================
    # 5) 构建知识图谱
    # =========================
    build_knowledge_graph_from_parsed_result(parsed_result)
    print("知识图谱构建完成！")
```
